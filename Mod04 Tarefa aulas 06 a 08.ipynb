````python
# 1) link da API
url_exercicio = 'https://apidatalake.tesouro.gov.br/ords/sadipem/tt/pvl'
````

````python
# 2) método get e validação do link
r = requests.get(url_exercicio)
r.status_code
````

````python
# 3) arquivo json com os dados do COPEM
Copem_exercicio_json = r.json()
````

````python
# 4) DataFrame com os dados do COPEM
Copem_df_exercicio = pd.DataFrame(Copem_exercicio_json['items'])
````

````python
# 5) cinco (5) primeiras linhas
Copem_df_exercicio.head(5)
````

````python
# 6) DataFrame com os dados do COPEM para municípios do estado do Rio de Janeiro (RJ)
Copem_df_RJ = Copem_df_exercicio.query('tipo_interessado == "Município" & uf == "RJ"')
Copem_df_RJ.head(5)
````

````python
# 7) Tipo e # do status de solicitação
Copem_df_RJ['status'].value_counts()
````

````python
# 8) Ano da coluna 'data_status'
data_status_df = Copem_df_RJ['data_status'].str.split("/", n = 2, expand = True)
ano_data_status = data_status_df[2]

# frequência dos anos
ano_data_status.value_counts()
````

Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

````python
# 1) Seu código aqui
# para o df do exercício
# parâmetros da função
uf = 'MG'
tipo_interessado = 'Município'

def consulta_dados_exercicio(df, uf, tipo_interessado):
    df_filtro_exercicio = df[(df['uf'] == uf) & (df['tipo_interessado'] == tipo_interessado)]
    return df_filtro_exercicio

consulta_dados_exercicio(Copem_df_exercicio, uf, tipo_interessado)
````

````python
# 2) Seu código aqui
# Dataframe com uf = MG e status = 'arquivado por decurso de prazo'
Copem_df_estado_mg_status = Copem_df_exercicio[(Copem_df_exercicio['uf'] == 'MG') & (Copem_df_exercicio['status'] == 'Arquivado por decurso de prazo')]

# contagem de linhas que atendem os critérios acima
Copem_df_estado_mg_status['id_pleito'].nunique()
````

````python
# 3) Seu código aqui
# filtro uf = 'BA'
Copem_df_ba = Copem_df_exercicio[(Copem_df_exercicio['uf'] == 'BA')]

# lista com todos os status
Copem_df_ba['status'].unique()

# filtro por meio do método contains para todos os status que possuem o termo 'Deferido'
Copem_df_ba_status_deferido = Copem_df_ba[Copem_df_ba['status'].str.contains('Deferido')]

# frequência dos municipios que contém status = 'Deferido'
Copem_df_ba_status_deferido['interessado'].value_counts()
````

````python
# 4) Seu código aqui
# filtro Dataframne
Copem_df_ba_estado = Copem_df_exercicio[(Copem_df_exercicio['uf'] == 'BA') & (Copem_df_exercicio['tipo_interessado'] == 'Estado')]

# salvando o Dataframe num arquivo .csv
Copem_df_ba_estado.to_csv(r'Copem_df_ba_estado.csv', index = False)
````
